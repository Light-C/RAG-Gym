```python
import os
import sys

# 设置 Java 21 环境变量
os.environ['JAVA_HOME'] = '/data/jialiang/jdk-21'
os.environ['PATH'] = os.environ['JAVA_HOME'] + '/bin:' + os.environ['PATH']

# 原有代码
sys.path.append(".")
import rag_gym

# 验证 Java 是否生效
print("Java Version:")
os.system("java -version")
```

```python
env = rag_gym.make(retriever_name = "BGE-M3", corpus_name = "Wikipedia", max_iter = 10, k = 32, rrf_k = 60, cache = False, HNSW = True) # `cache = True` will result in slower loading but faster retrieval
```

```python
agent = rag_gym.ReSearchAgent(llm_name = "/Qwen3-Next-80B-A3B-Instruct-int4-mixed-AutoRound", api = True, reward_llm_name = None, rag_llm_name = "/Qwen3-Next-80B-A3B-Instruct-int4-mixed-AutoRound")
```

```python
question = "What was the father of the last surviving Canadian father of Confederation?"
observation, info = env.reset(question = question)
```

```python
max_iterations = 10
temperature = 0.0
n_actions = 1
for i in range(max_iterations):
    print(f"[Time step {i}]")
    print(f"Generating {n_actions} candidate action(s)...")
    action = agent.generate_action(
        state = observation,
        temperature = temperature,
        num_actions = n_actions,
    )[0]
    print(f"Action taken: {action.return_as_json()}")
    observation, reward, terminated, truncated, info = env.step(action)
    if terminated or truncated:
        break
    print("")

print(f"Cache of retrieved documents:\n\n{observation.history.return_as_json(return_documents=True)}")
# print(f"Cache of summarized answers:\n\n{agent.rag_module.qa_cache}")
```
